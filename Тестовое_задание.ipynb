```python
#скачиваем основные необходимые библиотеки
import pandas as pd
import numpy as np
```

```python
#считываем полученный файл
df = pd.read_csv("test_data.csv")

#проверяем корректность считывания 
df.head(3)
```

```python
#группируем значения по колонке 'dlg_id'
groups = list(df.groupby('dlg_id').groups)
groupping = df.groupby('dlg_id').groups
```

```python
#количество групп
groups_count = len(groups)
```

```python
#проходимся по каждой группе
for i in range(groups_count):
    print(f'\033[1mdlg_id группы: {i}\033[0m')
    #список требований к менеджеру
    requirements = list()
    #получаем группу
    number = list(groupping[groups[i]])
    #получаем 'размер' рассматриваемой группы
    counter = 0
    for item in number:
        counter+=1
    
    for n in number:
        role = df['role'][n]
        if role == 'manager':
        #print(df['text'][n])
            proverka = df['text'][n]
            proverka = proverka.lower()
            if 'здравствуйте' in proverka:
                requirements.append('поздоровался')
                print(f'Менеджер поздоровался. Он сказал: "{proverka}"')
            
            if 'доброе утро' in proverka:
                requirements.append('поздоровался')
                print(f'Менеджер поздоровался. Он сказал: "{proverka}"')
            
            if 'добрый день' in proverka:
                requirements.append('поздоровался')
                print(f'Менеджер поздоровался. Он сказал: "{proverka}"')
            
            if 'добрый вечер' in proverka:
                requirements.append('поздоровался')
                print(f'Менеджер поздоровался. Он сказал: "{proverka}"')
        
            if ('меня' and 'зовут') in proverka:
                print(f'Менеджер представился. Он сказал: "{proverka}"')
                trigger = 'меня'
                proverka = proverka.split()
                for i in range(len(proverka)):
                    if trigger in proverka[i]:
                        if proverka[i+1] == 'зовут':
                            print(f'Имя менеджера: {proverka[i+2]}')
                        else:
                            print(f'Имя менеджера: {proverka[i+1]}')
                
            if 'мое имя' in proverka:
                print(f'Менеджер представился. Он сказал: "{proverka}"')
                trigger = 'имя'
                proverka = proverka.split()
                for i in range(len(proverka)):
                    if trigger in proverka[i]:
                        if proverka[i+1] == 'зовут':
                            print(f'Имя менеджера: {proverka[i+2]}')
                        else:
                            print(f'Имя менеджера: {proverka[i+1]}')
            
            if 'да это' in proverka:
                print(f'Менеджер представился. Он сказал: "{proverka}"')
                trigger = 'это'
                proverka = proverka.split()
                for i in range(len(proverka)):
                    if trigger in proverka[i]:
                        if proverka[i+1] == 'зовут':
                            print(f'Имя менеджера: {proverka[i+2]}')
                        else:
                            print(f'Имя менеджера: {proverka[i+1]}')
            
            if 'компания' in proverka:
                trigger = 'компания'
                for i in range(len(proverka)):
                    if trigger in proverka[i]:
                        if proverka[i+2] == 'бизнес':
                            print(f'Название компании: {proverka[i+1]} {proverka[i+2]}')
                        else:
                            print(f'Название компании: {proverka[i+1]}')
                            
            if 'до свидания' in proverka:
                requirements.append('попрощался')
                print(f'Менеджер попрощался. Он сказал: "{proverka}"')
            
            if ('желаю вам хорошего' or 'хорошего') in proverka:
                requirements.append('попрощался')
                print(f'Менеджер попрощался. Он сказал: "{proverka}"')
            
            if 'всего доброго' in proverka:
                requirements.append('попрощался')
                print(f'Менеджер попрощался. Он сказал: "{proverka}"')
            
            else:
                continue
        
        else:
            continue

    if 'поздоровался' in requirements and 'попрощался' not in requirements:
        print('\033[1mВывод\033[0m: менеджер поздоровался с клиентом, но не попрощался\n')

    if 'попрощался' in requirements and 'поздоровался' not in requirements:
        print('\033[1mВывод\033[0m: менеджер попрощался с клиентом, но не поздоровался\n')

    if 'попрощался' in requirements and 'поздоровался' in requirements:
        print('\033[1mВывод\033[0m: менеджер выполнил все требования, он поздоровался и попрощался с клиеном\n')    
```
